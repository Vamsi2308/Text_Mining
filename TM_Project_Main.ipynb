```python
## By hand - Find and open this file on your computer.
## Once you confirm that it contains what you expect
## close it.


## Open the file for append
MyFILE=open(filename, "a")

## Go through the json text:
for items in jsontxt2["articles"]:
    print(items, "\n\n\n")

    ## How do I know to use "articles" here?
    ## Answer: I don't! YOU must look at the json text
    ## to see the names of the categories. 
    ## When using newsapi.org and json output, one of the
    ## json categories is "articles". One is "source", etc...
    ## This is DIFFERENT for each API you use.
## Take a look at Figure 5 above and you will see what I mean
## You will see "articles" and "source" etc. and the hierarchy
## JSON tutorial
## https://www.w3schools.com/js/js_json_intro.asp


    Source=items["source"]["id"]
    print(Source)
    
    Date=items["publishedAt"]
    ##clean up the date
    NewDate=Date.split("T")
    Date=NewDate[0]
    print(Date)
    
#     Author=items["author"]
#     print(Author)
    
#     ## CLEAN the Author
#     ##----------------------------------------------------------
#     ##Replace punctuation with space
#     # Accept one or more copies of punctuation         
#     # plus zero or more copies of a space
#     # and replace it with a single space
#     Author=items["author"]
#     Author=re.sub(r'[,.;@#?!&$\-\']+', ' ', Author, flags=re.IGNORECASE)
#     Author=re.sub(' +', ' ', Author, flags=re.IGNORECASE)
#     Author=re.sub(r'\"', ' ', Author, flags=re.IGNORECASE)
    
#     # and replace it with a single space
#     ## NOTE: Using the "^" on the inside of the [] means
#     ## we want to look for any chars NOT a-z or A-Z and replace
#     ## them with blank. This removes chars that should not be there.
#     Author=re.sub(r'[^a-zA-Z]', " ", Author, flags=re.VERBOSE)
#     Author=Author.replace(',', '')
#     Author=' '.join(Author.split())
#     Author=re.sub("\n|\r", "", Author)
    
    ## CLEAN the Title
    ##----------------------------------------------------------
    ##Replace punctuation with space
    # Accept one or more copies of punctuation         
    # plus zero or more copies of a space
    # and replace it with a single space
    Title=items["title"]
    Title=re.sub(r'[,.;@#?!&$\-\']+', ' ', Title, flags=re.IGNORECASE)
    Title=re.sub(' +', ' ', Title, flags=re.IGNORECASE)
    Title=re.sub(r'\"', ' ', Title, flags=re.IGNORECASE)
    
    # and replace it with a single space
    ## NOTE: Using the "^" on the inside of the [] means
    ## we want to look for any chars NOT a-z or A-Z and replace
    ## them with blank. This removes chars that should not be there.
    Title=re.sub(r'[^a-zA-Z]', " ", Title, flags=re.VERBOSE)
    Title=Title.replace(',', '')
    Title=' '.join(Title.split())
    Title=re.sub("\n|\r", "", Title)
    
    ## CLEAN the Description
    ##----------------------------------------------------------
    
    Headline=items["description"]
    Headline=re.sub(r'[,.;@#?!&$\-\']+', ' ', Headline, flags=re.IGNORECASE)
    Headline=re.sub(' +', ' ', Headline, flags=re.IGNORECASE)
    Headline=re.sub(r'\"', ' ', Headline, flags=re.IGNORECASE)
    Headline=re.sub(r'[^a-zA-Z]', " ", Headline, flags=re.VERBOSE)
    ## Be sure there are no commas in the headlines or it will
    ## write poorly to a csv file....
    Headline=Headline.replace(',', '')
    Headline=' '.join(Headline.split())
    Headline=re.sub("\n|\r", "", Headline)
    
    ### AS AN OPTION - remove words of a given length............
    Headline = ' '.join([wd for wd in Headline.split() if len(wd)>3])

    #print("Author: ", Author, "\n")
    #print("Title: ", Title, "\n")
    #print("Headline News Item: ", Headline, "\n\n")
    
    #print(Author)
    print(Title)
    print(Headline)
    
    WriteThis=str(Date)+","+str(Source)+","+ str(Title) + "," + str(Headline) + "\n"
    
    MyFILE.write(WriteThis)
    
## CLOSE THE FILE
MyFILE.close()

## The output looks similar to this - just with different news ;)
##Author:  BBC News 

##Title:  Pope Francis addresses violence against women on Colombia visit 

##Headline News Item:  Pope Francis calls for respect for 
##"strong and influential" women during a five-day trip to Colombia. 
##--------------------------------------------------------
#FYI    
#do = jsontxt['articles'][0]["author"]
#print(do)

############### PROCESS THE FILE ######################
# , error_bad_lines=False
## Read to DF
BBC_DF=pd.read_csv(filename)
print(BBC_DF.head())
# iterating the columns 
for col in BBC_DF.columns: 
    print(col) 
    
print(BBC_DF["Headline"])

## REMOVE any rows with NaN in them
BBC_DF = BBC_DF.dropna()
print(BBC_DF["Headline"])
print(BBC_DF.shape)

### Tokenize and Vectorize the Headlines
## Create the list of headlines
HeadlineLIST=[]
for next1 in BBC_DF["Headline"]:
    HeadlineLIST.append(next1)

print("The headline list is")
print(HeadlineLIST)
```

```python
# Initialize NLTK stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Perform stemming and lemmatization
stemmed_text = [' '.join([stemmer.stem(word) for word in nltk.word_tokenize(sentence)]) for sentence in HeadlineLIST]
lemmatized_text = [' '.join([lemmatizer.lemmatize(word) for word in nltk.word_tokenize(sentence)]) for sentence in HeadlineLIST]
```

```python
# Initialize CountVectorizer stemmed
my_count_vectorizer_stemmed = CountVectorizer(input="content", lowercase=True, stop_words = "english")

# Convert text data to matrices of token counts using CountVectorizer
count_matrix_stemmed = my_count_vectorizer_stemmed.fit_transform(stemmed_text)  # create a sparse matrix
print("Type of CV matrix (stemmed):", type(count_matrix_stemmed))

#vocab is a vocabulary list
vocab_stemmed_cv = my_count_vectorizer_stemmed.get_feature_names()  # change to a list
print("Vocabulary list (stemmed):", list(vocab_stemmed_cv)[10:20])

# Convert the sparse matrix to a regular array
count_matrix_stemmed = count_matrix_stemmed.toarray()  # convert to a regular array
print(type(count_matrix_stemmed))

# Get column names
ColumnNames_stemmed=my_count_vectorizer_stemmed.get_feature_names()

# Create a DataFrame from the CountVectorizer output
cv_df_stemed=pd.DataFrame(count_matrix_stemmed,columns=ColumnNames_stemmed)
print(cv_df_stemed)

cv_df_stemed.to_csv('cv_df_stemmed.csv', index=False)

###################-----------------------------------

# Initialize CountVectorizer lemmatized
my_count_vectorizer_lemmatized = CountVectorizer(input="content", lowercase=True, stop_words = "english")

# Convert text data to matrices of token counts using CountVectorizer
count_matrix_lemmatized = my_count_vectorizer_lemmatized.fit_transform(lemmatized_text)  # create a sparse matrix
print("Type of CV matrix (lemmatized):",type(count_matrix_lemmatized))

#vocab is a vocabulary list
vocab_lemmatized_cv = my_count_vectorizer_lemmatized.get_feature_names()  # change to a list
print("Vocabulary list (lemmatized):", list(vocab_lemmatized_cv)[10:20])

# Convert the sparse matrix to a regular array
count_matrix_lemmatized = count_matrix_lemmatized.toarray()  # convert to a regular array
print(type(count_matrix_lemmatized))

# Get column names
ColumnNames_lemmatized=my_count_vectorizer_lemmatized.get_feature_names()

# Create a DataFrame from the CountVectorizer output
cv_df_lemmatized=pd.DataFrame(count_matrix_lemmatized,columns=ColumnNames_lemmatized)
print(cv_df_stemed)

cv_df_lemmatized.to_csv('cv_df_lemmatized.csv', index=False)
```

```python
# Initialize TfidfVectorizer Stemmed
tfidf_vectorizer_stemmed = TfidfVectorizer(input="content", lowercase=True, stop_words="english")

# Convert text data to matrices of TF-IDF features using TfidfVectorizer
tfidf_matrix_stemmed = tfidf_vectorizer_stemmed.fit_transform(stemmed_text)
print(type(tfidf_matrix_stemmed))

# Get the vocabulary list
vocab_stemmed_tfidf = tfidf_vectorizer_stemmed.get_feature_names()
print(list(vocab_stemmed_tfidf)[10:20])

# Convert the sparse matrix to a regular array
tfidf_matrix_stemmed = tfidf_matrix_stemmed.toarray()
print(type(tfidf_matrix_stemmed))

# Get column names
column_names_stemmed = tfidf_vectorizer_stemmed.get_feature_names()

# Create a DataFrame from the TfidfVectorizer output
tfidf_df_stemmed = pd.DataFrame(tfidf_matrix_stemmed, columns=column_names_stemmed)
print(tfidf_df_stemmed)

tfidf_df_stemmed.to_csv('tfidf_df_stemmed.csv', index=False)

###################-------------

# Initialize TfidfVectorizer lemmatized
tfidf_vectorizer_lemmatized = TfidfVectorizer(input="content", lowercase=True, stop_words="english")

# Convert text data to matrices of TF-IDF features using TfidfVectorizer
tfidf_matrix_lemmatized = tfidf_vectorizer_lemmatized.fit_transform(lemmatized_text)
print(type(tfidf_matrix_lemmatized))

# Get the vocabulary list
vocab_lemmatized_tfidf = tfidf_vectorizer_lemmatized.get_feature_names()
print(list(vocab_lemmatized_tfidf)[10:20])

# Convert the sparse matrix to a regular array
tfidf_matrix_lemmatized = tfidf_matrix_lemmatized.toarray()
print(type(tfidf_matrix_lemmatized))

# Get column names
column_names_lemmatized = tfidf_vectorizer_lemmatized.get_feature_names()

# Create a DataFrame from the TfidfVectorizer output
tfidf_df_lemmatized = pd.DataFrame(tfidf_matrix_lemmatized, columns=column_names_lemmatized)
print(tfidf_df_lemmatized)

tfidf_df_lemmatized.to_csv('tfidf_df_lemmatized.csv', index=False)
```
